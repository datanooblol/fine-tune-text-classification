```python
import re
from pythainlp.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer

def mixed_thai_english_tokenizer(text):
    tokens = []
    # Split by spaces first
    for segment in text.split():
        # Check if segment contains Thai characters
        if re.search(r'[\u0E00-\u0E7F]', segment):
            # Tokenize Thai part
            tokens.extend(word_tokenize(segment, engine='newmm'))
        else:
            # Keep English/technical terms as-is (lowercased)
            tokens.append(segment.lower())
    return tokens

# Use with TF-IDF
vectorizer = TfidfVectorizer(
    tokenizer=mixed_thai_english_tokenizer,
    max_features=1000,
    ngram_range=(1, 2),
)

# Example
text = "โปรโมชั่นสำหรับลูกค้า iPhone 15 Pro ลดราคา 50% คลิกเลย API integration ฟรี"
# Output: ['โปรโมชั่น', 'iphone', '15', 'pro', 'ลดราคา', '50%', 'คลิก', 'เลย', 'api', 'integration', 'ฟรี']
```

```python
tokens = vectorizer.fit_transform([text])
tokens
```
