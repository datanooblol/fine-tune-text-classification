```python
from package.utils import load_keywords

config = load_keywords(config_path="./keywords.config.yaml")
keywords = config["keywords"]
keywords
```

```python
texts = [
    "This is a sample text containing machine learning and artificial intelligence.",
    "Another text about deep learning and neural networks.",
    "This is clear and concise text without any technical terms.",
    "machne lerning is cool",
    "artifical inteligence",
    "text"
    
]
```

```python
from package.utils import contains_keyword
for text in texts:
    print(contains_keyword(text, keywords))
```

```python
from package.utils import contains_keyword_fuzzy
for text in texts:
    print(contains_keyword_fuzzy(text, keywords, threshold=80))
```
